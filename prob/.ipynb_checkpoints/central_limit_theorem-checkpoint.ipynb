```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def dist_plotter(s):
    sns.distplot(s, hist=True, kde=False, 
             bins=int(180/5), color = 'blue',
             hist_kws={'edgecolor':'black'})
```

```python
N = 1000
M = 1000
```

# Bernoulli

```python
# Bernoulli distrobution
p = 0.5
samples = [(int)(np.random.uniform(0,1,1)[0] > p) for _ in range(0, M)]
dist_plotter(samples)
```

```python
# Mean distribution
avg_samples= []
for _ in range(0, M):
    avg_samples.append(np.mean([(int)(np.random.uniform(0,1,1)[0] > p) for _ in range(0, M)]))
dist_plotter(avg_samples)
```

```python
# Median distribution
median_samples = []
for _ in range(0, M):
    median_samples.append(np.median([(int)(np.random.uniform(0,1,1)[0] > p) for _ in range(0, M)]))
dist_plotter(median_samples)
```

# Uniform

```python
# Uniform distrobution
samples = np.random.uniform(0,1,N)
dist_plotter(samples)
```

```python
avg_samples = [np.random.uniform(0,1,N).mean() for _ in range(0, M)]
dist_plotter(avg_samples)
```

```python
med_samples = [np.median(np.random.uniform(0,1,N)) for _ in range(0, M)]
dist_plotter(med_samples)
```

# Normal distribution


```python
samples = np.random.normal(0.5, 1, N)
dist_plotter(samples)
```

```python
avg_samples = [np.random.normal(0.5, 1, N).mean() for _ in range(0, M)]
dist_plotter(avg_samples)
```

```python
# Exponential
# Normal distribution
samples = np.random.exponential(1.0, N)
dist_plotter(samples)
```

```python
avg_samples = [np.random.exponential(1.0, N).mean() for _ in range(0, M)]
dist_plotter(avg_samples)
```

```python
med_samples = [np.median(np.random.exponential(1.,N)) for _ in range(0, M)]
dist_plotter(med_samples)
```
